# Descritores

<div style="text-align:justify">
Este notebook explica a ideia por trás dos descritores de dados e como a representação vetorial é essencial para o aprendizado de máquina. Vamos explorar os dados em diferentes dimensionalidades e ver como isso impacta na interpretação.
</div>

### Carregando Pacotes

```python
import pandas, numpy
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

### Configurações

```python
pyplot.rcParams['figure.figsize'] = (14, 8)
```

### Dataset

```python
df = pandas.DataFrame(data={
    'id': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'age': [19, 28, 21, 32, 33, 20, 22, 31, 20],
    'gender': [0, 0, 0, 0, 1, 1, 1, 1, 0],
    'salary': [1100, 3000, 2000, 3100, 3150, 1250, 1100, 2950, 1050]
})

df.head()
```

### Features

```python
x = df['age'].values
y = df['salary'].values
z = df['gender'].values

col = ['red' if c == 0 else 'blue' for c in z]
```

### Plot 2D

<div style="text-align:justify">
Analisando os dados pelas dimensões de <code>idade</code> e <code>salário</code> é possível inferir que existe uma correlação entre os dois dados, de modo que quando maior a idade, maior é o salário. Uma hipótese para esse acontecimento é que as pessoas mais velhas são mais experiêntes em suas funções.
</div>

```python
pyplot.scatter(x, y)
pyplot.show()
```

<div style="text-align:justify">
Claramente existem exceções. É possível notar uma pessoa de 21 anos com salário acima da média, comparado a outras pessoas da sua idade.
</div>

### Plot 3D

<div style="text-align:justify">
Vamos agora incluir uma terceira dimensão na análise dos vetores.
</div>

```python
ax = Axes3D(pyplot.figure())
for nx, ny, nz, c in zip(x, y, z, col):
    ax.scatter(nx, ny, nz, color=c)
    
pyplot.show()
```

### 

### Representação Vetorial

```python
for nx, ny, nz in zip(x, y, z):
    print('[{} {} {}]'.format(nx, ny, nz))
```
